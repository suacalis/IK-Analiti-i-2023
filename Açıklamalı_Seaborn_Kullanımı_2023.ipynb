#Seaborn'a Giriş

```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/suacalis/IK-Analitigi-2023/main/ikanalitikanket.csv',sep=';')
```

```python
df.info()
```

Countplot Grafiği

Pandas'da value_counts komutu ile değer olarak bulduğumuz verileri countplot grafik olarak çizmemize yardımcı olur

```python
sns.countplot(x='Cinsiyet', data=df);
```

```python
plt.figure(figsize=(14,6))
sns.countplot(x='Doğum_İl', data=df);
```

```python
sns.countplot(x='Doğum_Yılınız', data=df);
```

```python
sns.countplot(x='Ayakkabı Numaranız', data=df);
 
# X eksenindeki etiketleri döndür
plt.xticks(rotation=-45)
```

Pairplot Grafiği

Sayısal değişkenler arasındaki ilişkiyi matris formatında göstermeye yarar.

Aşağıdaki grafikte Öğretim Türü ile diğer değişkenler arasındaki ilişkiyi gösterir.

```python
sns.pairplot(df, hue='Cinsiyet');
```

```python
sns.pairplot(df,hue='Baba_Eğitim');
```

Box Plot (Kutu Grafiği)

Genellikle gruplar arasında bir veri dağılımını göstermenin görsel bir temsilidir.

Box plot beş parçadan oluşur
minimum

ilk çeyrek

medyan (ikinci çeyrek)

üçüncü çeyrek

maksimum

Kutu ne kadar uzun olursa veri o kadar dağılmış olur. Kutu ne kadar küçük olursa veri o kadar az dağılmış olur. Bir kutu grafiğini incelerken, bir aykırı değer, kutu grafiğinin bıyıklarının (dış tarafta kalan çizgiler) dışında bulunan bir veri noktası olarak tanımlanabilir.

```python
plt.figure(figsize=(9,4))
sns.boxplot(y=df['Not Ortalamanız'],labels=['Not Ortalamanız'], data=df)
plt.title("Not Ortalamanız")
plt.xlabel("Not Ortalamanız")
plt.show()
```

```python
sns.boxplot(x="Öğretim_Türünüz", y="Not Ortalamanız", data=df);
```

```python
#hue komutu grafiği farklı değişkenlere göre sınıflandırmamıza yardımcı olur.
sns.boxplot(x="Öğretim_Türünüz", y="Not Ortalamanız",hue="Cinsiyetiniz",data=df);
```

```python
#Violinplot boxplot'la benzerlik gösterir. Grafiğin şekli farklılık gösterir.
sns.violinplot(x="Öğretim_Türünüz", y="Not Ortalamanız", hue="Cinsiyetiniz", data=df);
```

```python
sns.violinplot(x="Öğretim_Türünüz", y="Not Ortalamanız", hue="Cinsiyetiniz", data=df, split=True);
```

```python
df.corr()
```

Heatmap (Isı Haritası)

df.corr() ile bulduğumuz korelasyon değerlerini görselleştirmek için kullanılır. 

Daha yaygın değerleri veya daha yüksek etkinlikleri temsil etmek için daha parlak renkler kullanılır daha az yaygın veya etkinlik değerlerini temsil etmek için daha koyu renkler tercih edilir.


```python
plt.figure(figsize=(14,6))
corr = df.corr()
sns.heatmap((corr),annot=True,cmap = 'Blues' );
```

Histogram Grafiği

Seaborn içerisindeki, dağılımı görselleştirmek için kullanılan distplot() fonksiyonu, tek değişkenli bir dağılıma hızlıca göz atmamıza yardımcı olur.

```python
sns.distplot(df['Not Ortalamanız']);
```

```python
#Seaborn'un yeni sürümlerinde distplot() fonksiyonun yerini displot alacaktır.
sns.displot(df['Not Ortalamanız'])
```

```python
sns.displot(df['Boyunuz ( cinsinden)']);
```

```python
sns.histplot(df['Boyunuz ( cinsinden)']);
```

Lmplot Grafiği

lmplot() fonksiyonu ile regresyon çizgisini oturtmak ve grafiğini çıkartmak için kullanıyoruz.

```python
sns.lmplot(x='Kilonuz', y='Ayakkabı Numaranız', data=df);
```

```python
sns.lmplot(x='Boyunuz ( cinsinden)', y='Ayakkabı Numaranız', data=df)
```

```python
sns.lmplot(x = 'Kilonuz', y = 'Ayakkabı Numaranız', fit_reg = True, hue = 'Cinsiyetiniz', data = df);
```

```python
sns.lmplot(x = 'Boyunuz ( cinsinden)', y = 'Ayakkabı Numaranız', fit_reg = True, hue = 'Cinsiyetiniz', data = df);
```

Sütun (Bar) Grafiği

Sütun grafikler elimizdeki kategorik değişkenleri görselleştirmek için kullanılır.

```python
#Sütun Grafiği
#Sütun grafiği, kategorik değişkenleri görselleştirmede kullanılır.
plt.figure(figsize=(9,4))
sns.barplot(x=df['Doğum Yılınız'], y=df['Kilonuz'], data=df)
plt.title("Doğum Tarihine Göre Kilonuz")
plt.xlabel("Doğum Tarihi")
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.barplot(x=df['Doğum Yılınız'], y=df['Kilonuz'], data=df)
plt.title("Doğum Tarihine Göre Kilonuz")
plt.xlabel("Doğum Tarihi")
plt.ylabel("Kilo")
plt.show()
```

Scatter Plot

Bir Dağılım grafiği, esas olarak iki sayısal grup arasındaki ilişkiyi ve bu ilişkinin yönünü dağınık noktalar çizerek anlatır.

Grafikteki her nokta tek bir gözlemi temsil edecek şekilde gösterilir:
X ekseni örneğin bir özelliğini temsil eder
Y ekseni, aynı örneğin farklı bir özelliğini temsil eder.

Dağılım grafikleri, iki değişken arasındaki ilişkiyi veya korelasyonu göstermek için kullanışlıdır.


```python
plt.figure(figsize=(9,4))
sns.scatterplot(x=df['Doğum Yılınız'], y=df['Kilonuz'], data=df)
plt.title("Doğum Tarihine Göre Kilonuz")
plt.xlabel("Doğum Tarihi")
plt.ylabel("Puan")
plt.show()
```
