```python
import PyPDF2

# Open the PDF file in read-binary mode
with open('/content/1 Introducing Employment Relations tr.pdf', 'rb') as file:
    # Create a PDF object
    pdf = PyPDF2.PdfReader(file)

    # Get the number of pages in the PDF
    num_pages = len(pdf.pages)

    # Iterate through all the pages
    for i in range(num_pages):
        # Get the page object
        page = pdf.pages[i]

        # Extract the text from the page
        text = page.extract_text()

        # Print the text
        print(text)
```

```python
import PyPDF2

with open('/content/1 Introducing Employment Relations tr.pdf', 'rb') as file:
    pdf = PyPDF2.PdfReader(file)
    num_pages = len(pdf.pages)
    text = ''
    for i in range(num_pages):
        page = pdf.pages[i]
        text += page.extract_text()

# Write the extracted text to a file
with open('text.txt', 'w', encoding='utf-8') as file:
    file.write(text)
```

```python
import PyPDF2

# Create a new PDF object
pdf = PyPDF2.PdfWriter()

# Open a text file in read-only mode
with open('document.txt', 'r') as file:
  # Read the text from the file
  text = file.read()

# Create a new page
page = PyPDF2.pdf.PageObject.createBlankPage()

# Add the text to the page
page.text = text

# Add the page to the PDF
pdf.addPage(page)

# Open a new PDF file in write-binary mode
with open('new_document.pdf', 'wb') as file:
  # Write the PDF to the file
  pdf.write(file)
```

```python
import PyPDF2

# Open the PDF file in read-binary mode
with open('existing_document.pdf', 'rb') as file:
  # Create a PDF object
  pdf = PyPDF2.PdfFileReader(file)
  
  # Create a new page
  page = PyPDF2.pdf.PageObject.createBlankPage()
  page.text = "This is a new page."
  
  # Add the new page to the PDF
  pdf.addPage(page)
  
  # Create a new PDF file in write-binary mode
  with open('modified_document.pdf', 'wb') as out_file:
    # Write the modified PDF to the file
    pdf.write(out_file)
```

```python
import PyPDF2

# Open the PDF file in read-binary mode
with open('existing_document.pdf', 'rb') as file:
  # Create a PDF object
  pdf = PyPDF2.PdfFileReader(file)
  
  # Get the page object
  page = pdf.getPage(0)
  
  # Rotate the page by 90 degrees clockwise
  page.rotateClockwise(90)
  
  # Add the modified page to the PDF
  pdf.addPage(page)
  
  # Create a new PDF file in write-binary mode
  with open('modified_document.pdf', 'wb') as out_file:
    # Write the modified PDF to the file
    pdf.write(out_file)
```

```python
# Create a new page with the watermark text
watermark_page = PyPDF2.pdf.PageObject.createBlankPage()
watermark_page.text = "CONFIDENTIAL"

# Add the watermark page to the PDF as a layer above the existing pages
pdf.addPage(watermark_page, prepend=True)
```

```python
import PyPDF2

# Open the PDF file in read-binary mode
with open('existing_document.pdf', 'rb') as file:
  # Create a PDF object
  pdf = PyPDF2.PdfFileReader(file)
  
  # Load the certificate and private key
  with open('certificate.pem', 'rb') as cert_file:
    certificate = PyPDF2.pkcs7.load_pem_x509_certificate(cert_file.read())
  with open('private_key.pem', 'rb') as key_file:
    private_key = PyPDF2.pkcs7.load_pem_private_key(key_file.read())
    
  # Sign the PDF with the certificate and private key
  signature = PyPDF2.pkcs7.sign(pdf.document, private_key, certificate, ['data'])
  
  # Add the signature to the PDF
  pdf.document.addSignature(signature)
```
