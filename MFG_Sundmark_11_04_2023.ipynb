```python
value_counts=df['Gender'].value_counts()
ax=df.Gender.value_counts().plot.bar()
# Add annotations to the bar chart
for i, v in enumerate(value_counts):
    ax.text(i, v + 10, str(v), ha='center', fontweight='bold')
```

```python
import matplotlib.pyplot as plt

# Generate the value counts and plot the bar chart
value_counts = df['Gender'].value_counts()
ax = value_counts.plot(kind='bar')

# Add annotations to the bar chart
for i, v in enumerate(value_counts):
    ax.text(i, v + 10, str(v), ha='center', fontweight='bold')

# Set the axis labels and title
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
ax.set_title('Gender Distribution')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the value counts
value_counts = df['Gender'].value_counts()

# Create the bar plot using Seaborn
ax = sns.barplot(x=value_counts.index, y=value_counts.values)

# Add annotations to the bar plot
for i, v in enumerate(value_counts.values):
    ax.annotate(str(v), xy=(i, v), ha='center', va='bottom', fontsize=12)

# Set the axis labels and title
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
ax.set_title('Gender Distribution')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the value counts
value_counts = df['Gender'].value_counts()

# Create a dictionary to map the original labels to the new labels
gender_map = {'M': 'Male', 'F': 'Female'}

# Map the original labels to the new labels
value_counts.index = value_counts.index.map(gender_map)

# Create the bar plot using Seaborn
ax = sns.barplot(x=value_counts.index, y=value_counts.values)

# Add annotations to the bar plot
for i, v in enumerate(value_counts.values):
    ax.annotate(str(v), xy=(i, v), ha='center', va='bottom', fontsize=12)

# Set the axis labels and title
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
ax.set_title('Gender Distribution')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size
fig, ax = plt.subplots(figsize=(3, 6))

# Generate the value counts
value_counts = df['Gender'].value_counts()

# Create a dictionary to map the original labels to the new labels
gender_map = {'M': 'Male', 'F': 'Female'}

# Map the original labels to the new labels
value_counts.index = value_counts.index.map(gender_map)

# Create the bar plot using Seaborn
sns.barplot(x=value_counts.index, y=value_counts.values, ax=ax)

# Add annotations to the bar plot
for i, v in enumerate(value_counts.values):
    ax.annotate(str(v), xy=(i, v), ha='center', va='bottom', fontsize=12)

# Set the axis labels and title
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
ax.set_title('Gender Distribution')

# Display the plot
plt.show()
```

```python
import plotly.express as px

# Generate the value counts
value_counts = df['Gender'].value_counts()

# Create the bar plot using Plotly Express
fig = px.bar(x=value_counts.index, y=value_counts.values, labels={'x': 'Gender', 'y': 'Count'})

# Add annotations to the bar plot
for i, v in enumerate(value_counts.values):
    fig.update_traces(texttemplate='%{y}', textposition='outside')
    fig.add_annotation(x=i, y=v, text=str(v), showarrow=False)

# Set the title
fig.update_layout(title_text='Gender Distribution')

# Display the plot
fig.show()
```
