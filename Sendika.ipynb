```python
df= df.sort_values(by=['Yıl', 'Plaka'])
```

```python
df=df.sort_values(by='Yıl', ascending=True)
```

```python
df.head()
```

```python
px.scatter(df, x="İşyeri Sayısı", y="Toplu İş Sözleşme Sayısı", animation_frame="Yıl", animation_group="İl",
           size="Üye İşçi Sayısı", color="İl", hover_name="İl",
           log_x = True,
           size_max=45, range_x=[100,30000], range_y=[0,4450])
```

```python
fig = px.scatter(df, x="Yıl", y="Toplu İş Sözleşme Sayısı", color='İl', size='İşçi Sayısı', size_max=20)
fig.show()
```
