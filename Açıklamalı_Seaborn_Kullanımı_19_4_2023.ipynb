```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks")

# Initialize the figure with a logarithmic x axis
f, ax = plt.subplots(figsize=(7, 6))
ax.set_xscale("log")

# Load the example planets dataset
planets = sns.load_dataset("planets")

# Plot the orbital period with horizontal boxes
sns.boxplot(x="distance", y="method", data=planets,
            whis=[0, 100], width=.6, palette="vlag")

# Add in points to show each observation
sns.stripplot(x="distance", y="method", data=planets,
              size=4, color=".3", linewidth=0)

# Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```

Box Plot (Kutu Grafiƒüi)

Genellikle gruplar arasƒ±nda bir veri daƒüƒ±lƒ±mƒ±nƒ± g√∂stermenin g√∂rsel bir temsilidir.

Box plot be≈ü par√ßadan olu≈üur
minimum

ilk √ßeyrek

medyan (ikinci √ßeyrek)

√º√ß√ºnc√º √ßeyrek

maksimum

Kutu ne kadar uzun olursa veri o kadar daƒüƒ±lmƒ±≈ü olur. Kutu ne kadar k√º√ß√ºk olursa veri o kadar az daƒüƒ±lmƒ±≈ü olur. Bir kutu grafiƒüini incelerken, bir aykƒ±rƒ± deƒüer, kutu grafiƒüinin bƒ±yƒ±klarƒ±nƒ±n (dƒ±≈ü tarafta kalan √ßizgiler) dƒ±≈üƒ±nda bulunan bir veri noktasƒ± olarak tanƒ±mlanabilir.

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['Not_Ortalama'],labels=['Not Ortalamanƒ±z'], data=df)
plt.title("Not Ortalamanƒ±z")
plt.xlabel("Not Ortalamanƒ±z")
plt.show()
```

S√ºtun (Bar) Grafiƒüi

S√ºtun grafikler elimizdeki kategorik deƒüi≈ükenleri g√∂rselle≈ütirmek i√ßin kullanƒ±lƒ±r.

```python
#S√ºtun Grafiƒüi
#S√ºtun grafiƒüi, kategorik deƒüi≈ükenleri g√∂rselle≈ütirmede kullanƒ±lƒ±r.
plt.figure(figsize=(9,4))
sns.barplot(x=df['Boy'], y=df['Ayakkabƒ±_Numara'], data=df)
plt.show()
```

Scatter Plot

Bir Daƒüƒ±lƒ±m grafiƒüi, esas olarak iki sayƒ±sal grup arasƒ±ndaki ili≈ükiyi ve bu ili≈ükinin y√∂n√ºn√º daƒüƒ±nƒ±k noktalar √ßizerek anlatƒ±r.

Grafikteki her nokta tek bir g√∂zlemi temsil edecek ≈üekilde g√∂sterilir:
X ekseni √∂rneƒüin bir √∂zelliƒüini temsil eder
Y ekseni, aynƒ± √∂rneƒüin farklƒ± bir √∂zelliƒüini temsil eder.

Daƒüƒ±lƒ±m grafikleri, iki deƒüi≈üken arasƒ±ndaki ili≈ükiyi veya korelasyonu g√∂stermek i√ßin kullanƒ±≈ülƒ±dƒ±r.


```python
plt.figure(figsize=(9,4))
sns.scatterplot(x=df['Boy'], y=df['Ayakkabƒ±_Numara'], data=df)
plt.title("Boya Ayakkabƒ± Numarasƒ±")
plt.xlabel("Boy")
plt.ylabel("Ayakkabƒ± Numarasƒ±")
plt.show()
```
